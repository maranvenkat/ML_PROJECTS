```python
retail=df.groupby(["Member_number","itemDescription"])["itemDescription"].count().unstack().fillna(0)

retail
```

```python
retail["berries"].value_counts()
```

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
    return 0
retail_basket=retail.applymap(encode)
```

```python
apr=apriori(retail_basket,min_support=0.07,use_colnames=True)
```

```python
df_rules=association_rules(apr,metric="lift",min_threshold=1)
df_rules
```

```python
def get_recommendations(product, rules, num_recommendations=5):
  recommendations=[]
  for idx, row in rules.iterrows():
    if product in row["antecedents"]:
      recommended_item = next(iter(row["consequents"]))
      recommendations.append((recommended_item, row["lift"]))
  recommendations.sort(key=lambda x: x[1],reverse=True)
  return recommendations[:num_recommendations]
```

```python
product_to_recommend="whole milk"
recommendations = get_recommendations(product_to_recommend,df_rules)
print(f"Recommendations for '{product_to_recommend}':")
for item, lift in recommendations:
  print(f"{item} (lift: {lift:.2f})")
```
