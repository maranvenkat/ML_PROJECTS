```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df=pd.read_excel("https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
#recency
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
current_date=df["InvoiceDate"].max()
recency_date=df.groupby("CustomerID")["InvoiceDate"].max().reset_index()
recency_date["Recency"]=(current_date-recency_date["InvoiceDate"]).dt.days

print(recency_date[["CustomerID","Recency"]].head())
```

```python
frequency_data=df.groupby("CustomerID")["InvoiceDate"].nunique().reset_index()
frequency_data.columns=("CustomerID","Frequency")

print(frequency_data.head())
```

```python
df["TotalPrice"]=df["Quantity"]*df["UnitPrice"]
```

```python
monetary_value=df.groupby("CustomerID")["TotalPrice"].sum().reset_index()
monetary_value.columns = ["CustomerID","Monetary"]

print(monetary_value.head())
```

```python
plt.figure(figsize=(10,8))
plt.hist(recency_date["Recency"],bins=50,color="skyblue",edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(frequency_data["Frequency"],bins=50,color="blue",edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(monetary_value["Monetary"],bins=50,color="lightcoral",edgecolor="black")
plt.show()
```

```python
rfm_df=pd.merge(recency_date,frequency_data,on="CustomerID")
rfm_df=pd.merge(rfm_df,monetary_value,on="CustomerID")

x=rfm_df[["Recency","Frequency","Monetary"]]

num_clusters=3

kmeans=KMeans(n_clusters=num_clusters,random_state=42)
rfm_df["Cluster"]=kmeans.fit_predict(x)

print(rfm_df[["CustomerID","Recency","Frequency","Monetary","Cluster"]].head())
```

```python
plt.figure(figsize=(10,8))
for cluster in range(num_clusters):
  cluster_data = rfm_df[rfm_df["Cluster"]==cluster]
  plt.scatter(cluster_data["Recency"],cluster_data["Frequency"],label=f"Cluster {cluster + 1}")

plt.title("RFM Cluster")
plt.legend()
plt.show()
```
